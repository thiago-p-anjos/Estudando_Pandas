### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros. Apresente também ao lado os estados na qual os maiores e menores preços foram registrados.

```python
preco_max = gasolina_comum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].max()
```

```python
preco_min = gasolina_comum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].min()
```

```python
id_preco_max = gasolina_comum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmax()
```

```python
id_preco_min = gasolina_comum.groupby('ANO-MES')['PREÇO MÉDIO REVENDA'].idxmin()
```

```python
estado_preco_max = gasolina_comum.loc[id_preco_max, ["ESTADO", "ANO-MES"]].set_index("ANO-MES")
```

```python
estado_preco_min = gasolina_comum.loc[id_preco_min, ["ESTADO", "ANO-MES"]].set_index("ANO-MES")
```

```python
dif_abs = preco_max - preco_min
```

```python
dif_perc = (preco_max- preco_min)/preco_min*100
```

```python
tab_desafio = pd.DataFrame()
```

```python
tab_desafio['preco_min'] = preco_min
```

```python
tab_desafio['preco_max'] = preco_max
```

```python
tab_desafio['dif_abs'] = dif_abs
```

```python
tab_desafio['dif_perc'] = dif_perc
```

```python
tab_desafio['estado_preco_max'] = estado_preco_max
```

```python
tab_desafio['estado_preco_min'] = estado_preco_min
```
